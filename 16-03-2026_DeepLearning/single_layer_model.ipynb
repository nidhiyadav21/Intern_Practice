```python
import tensorflow as tf
import numpy as np
import random
```

```python
tf.random.set_seed(42)
np.random.seed(42)
random.seed(42)
```

```python
x = np.array([-10, -5, -1, 1, 5, 10],dtype= float)
y = np.array([0, 0, 0, 1, 1, 1],dtype = float)
```

```python
model = tf.keras.Sequential([tf.keras.layers.Dense(units = 1, input_shape = [1],activation='sigmoid')])
```

```python
model.compile(optimizer= 'adam',loss = 'binary_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(x,y,epochs=50,verbose=1)
```

```python
test_val = np.array([7.0, -3.0])
predictions = model.predict(test_val)
```

```python
for idx, val in enumerate(test_val):
    # predictions[idx] will be an array like [probability], so we need [idx][0] for the scalar value
    category = "Positive" if predictions[idx][0] > 0.5 else "Negative"
    print(f"Number: {val} | Probability: {predictions[idx][0]:.4f} | Class: {category}")
```
